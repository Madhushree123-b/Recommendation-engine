```python
data=book_df.groupby(['userid','Book'])['Bookrating'].mean().reset_index()
data['Bookrating']=round(data['Bookrating'],0)
print(data.shape)
data.head()
```

```python
data.to_csv(r"D:\dataset\test.csv")
```

```python
user_data = data.pivot(index='userid',
                                 columns='Book',
                                 values='Bookrating').reset_index()
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_data.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_data = pd.DataFrame(user_sim)
```
